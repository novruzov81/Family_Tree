```python
import pandas as pd
kc_house_data = pd.read_csv('/content/kc_house_data.csv', sep=",", encoding='latin1')
kc_house_data.head()
```

```python
average_price = kc_house_data['price'].mean()
print("Средняя стоимость недвижимости:", average_price)
```

```python
import matplotlib.pyplot as plt
plt.hist(kc_house_data['sqft_living'], bins=100, facecolor='green')
plt.title('Распределение квадратуры жилой площади')
plt.xlabel('Квадратура жилой площади')
plt.ylabel('Количество объектов')
plt.show()
```

```python
plt.hist(kc_house_data['yr_built'], bins=100, facecolor='blue')
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество объектов')
plt.show()
```

```python
data = kc_house_data['waterfront'].value_counts()
data
plt.title('Распределение домов с видом на набережную')
names = data.index
values = data.values

plt.pie(values, labels=names, autopct='%.1f%%');
```

Вывод: 0,8% домов имеет вид на набережную


```python
data = kc_house_data['floors'].value_counts()
data
plt.title('Распределение домов по этажности')
names = data.index
values = data.values

plt.bar(names, values);
```

```python
data = kc_house_data['grade'].value_counts()
data
plt.title('Распределение состояния домов')
names = data.index
values = data.values

plt.bar(names, values);
```
